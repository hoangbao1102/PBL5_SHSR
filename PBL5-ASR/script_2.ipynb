```python
import os
import shutil

# ng dn ti th mc gc
root_dir = "data_cuted"

# ng dn ti cc th mc ch
train_dir = "data_train"
test_dir = "data_test"
val_dir = "data_val"

# To cc th mc ch nu cha tn ti
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Lp qua tng th mc con trong th mc gc
for folder in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder)

    # To cc th mc con trong cc th mc ch
    train_sub_dir = os.path.join(train_dir, folder)
    test_sub_dir = os.path.join(test_dir, folder)
    val_sub_dir = os.path.join(val_dir, folder)
    os.makedirs(train_sub_dir, exist_ok=True)
    os.makedirs(test_sub_dir, exist_ok=True)
    os.makedirs(val_sub_dir, exist_ok=True)

    # Ly danh sch cc tp tin trong th mc con
    files = os.listdir(folder_path)

    # S lng tp tin trong th mc con
    num_files = len(files)

    # Tnh s lng tp tin cn chia cho mi ln (6:2:2)
    num_train_files = int(0.6 * num_files)
    num_test_files = int(0.2 * num_files)
    num_val_files = int(0.2 * num_files)

    # Lp li qu trnh chia t l v sao chp 10 ln
    for _ in range(10):
        # To danh sch cc ch s tp tin
        file_indices = list(range(num_files))

        # Chia t l s lng tp tin
        train_indices = file_indices[:num_train_files]
        test_indices = file_indices[num_train_files:num_train_files + num_test_files]
        val_indices = file_indices[num_train_files + num_test_files:]

        # Lp qua cc ch s v sao chp tp tin vo cc th mc ch tng ng
        for index in train_indices:
            file = files[index]
            src = os.path.join(folder_path, file)
            dst = os.path.join(train_sub_dir, file)
            shutil.copyfile(src, dst)

        for index in test_indices:
            file = files[index]
            src = os.path.join(folder_path, file)
            dst = os.path.join(test_sub_dir, file)
            shutil.copyfile(src, dst)

        for index in val_indices:
            file = files[index]
            src = os.path.join(folder_path, file)
            dst = os.path.join(val_sub_dir, file)
            shutil.copyfile(src, dst)
```

```python
import os
import shutil

# ng dn ti th mc gc
root_dir = "data_cuted"

# ng dn ti cc th mc ch
train_dir = "data_train_1"
test_dir = "data_test_1"
val_dir = "data_val_1"

# To cc th mc ch nu cha tn ti
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Lp qua tng th mc con trong th mc gc
for folder in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder)

    # To cc th mc con trong cc th mc ch
    train_sub_dir = os.path.join(train_dir, folder)
    test_sub_dir = os.path.join(test_dir, folder)
    val_sub_dir = os.path.join(val_dir, folder)
    os.makedirs(train_sub_dir, exist_ok=True)
    os.makedirs(test_sub_dir, exist_ok=True)
    os.makedirs(val_sub_dir, exist_ok=True)

    # Ly danh sch cc tp tin trong th mc con
    files = os.listdir(folder_path)

    # S lng tp tin trong th mc con
    num_files = len(files)

    # S lng lnh ca mi ngi
    num_commands_per_person = int(num_files / 10)

    # Chia t l s lng lnh theo yu cu (6:2:2)
    num_train_files = int(num_commands_per_person * 0.6)
    num_test_files = int(num_commands_per_person * 0.2)
    num_val_files = int(num_commands_per_person * 0.2)

    # Lp li qu trnh chia t l v sao chp 10 ln
    for i in range(10):
        # To th mc con trong th mc ch
        test_sub_dir = os.path.join(test_dir, folder + str(i+1))
        val_sub_dir = os.path.join(val_dir, folder + str(i+1))
        os.makedirs(test_sub_dir, exist_ok=True)
        os.makedirs(val_sub_dir, exist_ok=True)

        # Tnh ch s bt u ca mi ngi trong danh sch tp tin
        start_index = i * num_commands_per_person

        # Chia t l s lng lnh
        train_indices = list(range(start_index, start_index + num_train_files))
        test_indices = list(range(start_index + num_train_files, start_index + num_train_files + num_test_files))
        val_indices = list(range(start_index + num_train_files + num_test_files, start_index + num_train_files + num_test_files + num_val_files))

        # Sao chp tp tin vo th mc ch tng ng
        for index in train_indices:
            file = files[index]
            src = os.path.join(folder_path, file)
            dst = os.path.join(train_sub_dir, file)
            shutil.copyfile(src, dst)

        for index in test_indices:
            file = files[index]
            src = os.path.join(folder_path, file)
            dst = os.path.join(test_sub_dir, file)
            shutil.copyfile(src, dst)

        for index in val_indices:
            file = files[index]
            src = os.path.join(folder_path, file)
            dst = os.path.join(val_sub_dir, file)
            shutil.copyfile(src, dst)
```

```python
import os
import shutil

def copy_all_commands(source_folders, destination_folder):
    # To th mc ch nu cha tn ti
    os.makedirs(destination_folder, exist_ok=True)

    # Lp qua tt c cc th mc ngun
    for source_folder in source_folders:
        # Lp qua tt c cc th mc con trong th mc ngun
        for root, dirs, files in os.walk(source_folder):
            # Lp qua tt c cc tp tin trong th mc con
            for file in files:
                # To ng dn y  n tp tin ngun
                source_path = os.path.join(root, file)

                # Sao chp tp tin vo th mc ch
                shutil.copy(source_path, destination_folder)

# Nhp ng dn ti cc th mc con
source_folders = []

folder='data_val_1/tatquat'
for i in range(1, 11):
    folder_path = folder
    source_folders.append(folder_path+str(i))

# Nhp ng dn ti th mc ch  sao chp cc lnh
destination_folder = 'data_val/tatquat'

# Sao chp tt c cc lnh t cc th mc con vo th mc ch
copy_all_commands(source_folders, destination_folder)

print(" sao chp thnh cng!")
```

```python
import os

def count_commands(folder_path):
    num_commands = 0
    for root, dirs, files in os.walk(folder_path):
        num_commands += len(files)
    return num_commands

# ng dn ti th mc data_test/xxx
folder_path = "data_train/batden"

# m s lng lnh trong th mc
num_commands = count_commands(folder_path)

print(f"S lng lnh trong {folder_path}: {num_commands}")
```

```python
import shutil

def delete_folders(source_folders):
    for folder in source_folders:
        # Xa th mc v ni dung bn trong
        shutil.rmtree(folder)
        print(f" xa th mc: {folder}")

# Nhp ng dn ti cc th mc con
source_folders = []
for i in range(1, 11):
    folder_path = 'data_test/batdieuhoa'
    source_folders.append(folder_path+str(i))

# Xa cc th mc trong source_folders
delete_folders(source_folders)
```

```python
import os
import librosa
import matplotlib.pyplot as plt

data_folder = 'data_show'

# Get the list of WAV files in the data folder
wav_files = [file for file in os.listdir(data_folder) if file.endswith('.wav')]

# Set the number of rows and columns for subplots
rows = 5
cols = 2

# Create the subplots with specified figsize
fig, axes = plt.subplots(rows, cols, figsize=(10, 12))

# Iterate over the WAV files
for i, wav_file in enumerate(wav_files[:rows * cols]):
    # Get the path to the WAV file
    wav_path = os.path.join(data_folder, wav_file)

    # Load the audio file
    audio, sr = librosa.load(wav_path)

    # Plot the waveform
    ax = axes[i // cols][i % cols]
    ax.plot(audio)
    ax.set_title(wav_file)

# Adjust the spacing between subplots
fig.subplots_adjust(hspace=0.5)

plt.savefig('wave_form_show')
# Show the plot
plt.show()
```
