```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Đọc tệp âm thanh
audio, sr = librosa.load('42_LVA_01_dongcua.wav')

# Trích xuất đặc trưng Mel Spectrogram
spec = librosa.feature.melspectrogram(y=audio, sr=sr)
spec_db = librosa.power_to_db(spec, ref=np.max)

# Vẽ đồ thị và lưu hình ảnh
plt.figure(figsize=(10, 4))
librosa.display.specshow(spec_db, sr=sr, x_axis='time', y_axis='mel', cmap='jet')
plt.axis('off')
plt.tight_layout()
plt.savefig('mel_spectrogram.png', bbox_inches='tight', pad_inches=0)
```
